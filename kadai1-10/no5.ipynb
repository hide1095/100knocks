```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import io

img_orig = io.imread('https://yoyoyo-yo.github.io/Gasyori100knock/assets/imori_256x256.png')

plt.imshow(img_orig)
plt.show()
```

```python
from matplotlib import cm
plt.figure(figsize=(12, 1))
plt.title('Hue')
for i in range(360):
    plt.vlines(i, 0, 1, color=cm.hsv(i / 360))
plt.show()
```

```python
hsv = cv2.cvtColor(img_orig, cv2.COLOR_RGB2HSV)
hsv[..., 0] = (hsv[..., 0] + 90) % 180
img_hsv = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
plt.subplot(1, 2, 1)
plt.title('')
plt.imshow(img_orig)
plt.subplot(1, 2, 2)
plt.title('')
plt.imshow(img_hsv)
plt.show()
```
