```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import io

img_orig = io.imread('https://yoyoyo-yo.github.io/Gasyori100knock/assets/imori_128x128.png ')

plt.imshow(img_orig)
plt.show()
```

```python
def blur(img):
    ans=img.copy()
    H,W,C = ans.shape
    BH=H//8
    BW=W//8
    for y in range(BH):
        for x in range(BW):
            for c in range(C):
                ans[y*8:(y+1)*8,x*8:(x+1)*8,c] = np.mean(ans[y*8:(y+1)*8,x*8:(x+1)*8,c].astype(np.int))
    
    return ans
```

```python
img_ans = blur(img_orig)
plt.subplot(1,2,1)
plt.imshow(img_orig)

plt.subplot(1,2,2)
plt.imshow(img_ans)

plt.show()
```
